# Linear Model Overfiting

## First set p=n


```python
import numpy as np
import random
import statsmodels.api as sm
import warnings
warnings.filterwarnings('ignore')
```

```python
# Set Seed
random.seed(10)
print(random.random())

n = 1000
p = n
```

```python
X = np.random.normal(0, 1, size=(n, p))
Y = np.random.normal(0, 1,n)
```

```python
mod = sm.OLS(Y, X)    # Describe model
res = mod.fit()
print(res.summary())
```

```python
print("p/n is")
print(p/n)
```

```python
print("R2 is")
res.rsquared
```

```python
print("Adjusted R2 is")
est2 = mod.fit()
est2.rsquared_adj
```

## Second, set p=n/2.

```python
random.seed(10)
n = 1000
p = n/2
```

```python
X = np.random.normal(0, 1, size=(n, int(p)))
Y = np.random.normal(0, 1,n)
mod = sm.OLS(Y, X)    # Describe model
res = mod.fit()
print(res.summary())
```

```python
print("p/n is")
print(p/n)

print("R2 is")
res.rsquared

print("Adjusted R2 is")
est2 = mod.fit()
est2.rsquared_adj
```

```python
print("p/n is \n",p/n )
#print("summary()\n",res.summary())
print("rsquared\n",est2.rsquared)
print("rsquared_adj\n",est2.rsquared_adj)
```
